```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df20=pd.read_table('chr_20_frequencies.frq',delim_whitespace=True)
df21=pd.read_table('chr_21_frequencies.frq',delim_whitespace=True)
df22=pd.read_table('chr_22_frequencies.frq',delim_whitespace=True)
```

```python
plt.figure(figsize=(15,4))
plt.hist(df20['MAF'])
plt.title('CHR20')
plt.show()

plt.figure(figsize=(15,4))
plt.hist(df21['MAF'],color='g')
plt.title('CHR21')
plt.ylim(0, 1000)
plt.show()

plt.figure(figsize=(15,4))
plt.hist(df22['MAF'],color='r')
plt.title('CHR22')
plt.show()
```
